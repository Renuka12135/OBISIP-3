```python
import speech_recognition as sr
import pyttsx3
import pywhatkit
import datetime

# Initialize text-to-speech engine
engine = pyttsx3.init()
engine.setProperty('rate', 150)  # Speaking speed

def speak(text):
    """Convert text to speech."""
    engine.say(text)
    engine.runAndWait()

def listen():
    """Capture voice input and convert to text."""
    recognizer = sr.Recognizer()
    with sr.Microphone() as source:
        recognizer.adjust_for_ambient_noise(source, duration=1)
        print("üé§ Listening... Please speak now.")
        try:
            # Timeout if user doesn't speak within 5 seconds; limit phrase to 5 seconds
            audio = recognizer.listen(source, timeout=5, phrase_time_limit=5)
            command = recognizer.recognize_google(audio)
            print(f"üó£Ô∏è You said: {command}")
            return command.lower()
        except sr.WaitTimeoutError:
            print("‚è±Ô∏è You didn‚Äôt speak in time.")
            speak("You didn‚Äôt speak in time. Try again.")
            return ""
        except sr.UnknownValueError:
            print("ü§∑ Sorry, I didn't catch that.")
            speak("Sorry, I didn't catch that.")
            return ""
        except sr.RequestError:
            print("üîå Speech service error. Please check your connection.")
            speak("Sorry, my speech service is down.")
            return ""

def respond(command):
    """Respond to specific voice commands."""
    if "hello" in command:
        speak("Hello! How can I assist you today?")
    elif "time" in command:
        current_time = datetime.datetime.now().strftime("%I:%M %p")
        print(f"üïí The time is {current_time}")
        speak(f"The time is {current_time}")
    elif "date" in command:
        current_date = datetime.datetime.now().strftime("%A, %B %d, %Y")
        print(f"üìÖ Today is {current_date}")
        speak(f"Today is {current_date}")
    elif "search" in command:
        query = command.replace("search", "").strip()
        print(f"üîç Searching for: {query}")
        speak(f"Searching for {query}")
        pywhatkit.search(query)
    elif command:
        print("ü§ñ Command not recognized.")
        speak("I'm not sure how to help with that. Try saying 'search', 'time', or 'date'.")

def main():
    """Main function to activate the assistant."""
    speak("Voice assistant activated. Say something...")
    print("üí¨ Voice assistant activated. Say 'exit' or 'stop' to quit.")
    while True:
        command = listen()
        if "exit" in command or "stop" in command:
            speak("Goodbye!")
            print("üëã Goodbye!")
            break
        elif command:
            respond(command)

# Run the assistant
if __name__ == "__main__":
    main()  
```
